Reemplazamos los datos faltantes por valores dependiendo de cada columna, así:
- ESTU_VALORMATRICULAUNIVERSIDAD = 'No sabe, no responde'
- FAMI_ESTRATOVIVIENDA = 'Sin Estrato'
- FAMI_TIENEINTERNET = 'No'
- FAMI_EDUCACIONPADRE = 'No Aplica'
- FAMI_EDUCACIONMADRE = 'No Aplica'
- ESTU_PAGOMATRICULAPROPIO = 'No sabe, no responde'


```python
df_train['ESTU_VALORMATRICULAUNIVERSIDAD'].fillna(value='No sabe, no responde', inplace=True)
df_train['FAMI_ESTRATOVIVIENDA'].fillna(value='Sin Estrato', inplace=True)
df_train['FAMI_TIENEINTERNET'].fillna(value='No', inplace=True)
df_train['FAMI_EDUCACIONPADRE'].fillna(value='No Aplica', inplace=True)
df_train['FAMI_EDUCACIONMADRE'].fillna(value='No Aplica', inplace=True)
df_train['ESTU_PAGOMATRICULAPROPIO'].fillna(value='No sabe, no responde', inplace=True)
```

Verificamos que haya quedado guardado

```python
df_train['ESTU_VALORMATRICULAUNIVERSIDAD'].unique()
df_train['FAMI_ESTRATOVIVIENDA'].unique()
df_train['FAMI_TIENEINTERNET'].unique()
df_train['FAMI_EDUCACIONPADRE'].unique()
df_train['FAMI_EDUCACIONMADRE'].unique()
df_train['ESTU_PAGOMATRICULAPROPIO'].unique()
```

Ahora vamos a mirar cuantos estudiantes hay por localidades.

```python
df_train.groupby('ESTU_PRGM_DEPARTAMENTO').count()['ESTU_PRGM_ACADEMICO']
```

Se muestra una gráfica con la cantidad de estudiantes por departamento, como era de esperarse, la mayoría de estudiantes son de los departamentos mas desarrolados, los cuales son en orden descendente: Bogotá, Antioquia, Valle, Atlantico y Santander.

```python
df_train.groupby('ESTU_PRGM_DEPARTAMENTO').count()['ESTU_PRGM_ACADEMICO'].plot(y='ESTU_PRGM_ACADEMICO', kind='bar')
```
